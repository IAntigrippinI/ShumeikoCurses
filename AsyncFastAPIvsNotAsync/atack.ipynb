```python
import asyncio
import aiohttp


async def get_data(id: int, endpoint: str):
    print(f"Начал выполнение: {id}")
    url = f"http://127.0.0.0:8000/{endpoint}/{id}"
    async with aiohttp.ClientSession() as Session:
        async with Session.get(url) as resp:
            print(f"Закончил выполнение: {id}")
```

```python
await get_data(1, 'sync')
```

```python
iter = 300
```

```python
await asyncio.gather(
    *[get_data(i, 'sync') for i in range(iter)]
)
```

```python
await asyncio.gather(
    *[get_data(i, 'async') for i in range(iter)]
)
```
